## Time Value of Money (TVM)

The TVM calculation of cash flow consist of the following quantities:

| Math Symbol | Explanation |
| --- | --- |
| $r$ | Nominal annual interest rate |
| $l$ | Number of years |
| $n$ | Number of payment period per year |
| $P$ | Payment amount |
| $PV$ | Present value of all payments (beginning of 1st year) |
| $FV$ | Future value of all payments (end of nth year) |

The mathematical relationship between the quantities is:

$$
\frac{FV}{(1+\frac{r}{n})^{nl}} = PV = \big( \frac{1 - (1+\frac{r}{n})^{-nl}}{\frac{r}{n}} \big) P
$$

Since we set the number of payments per year as 1. In the following usage of the calculator, we effectively regard "a year" as "a period". For example, a quarterly compounded 5 years loan with 4% nominal annual interest rate is effectively the same with yearly compounded 20 years loan with 1% interest rate in our calculation.

If we set the number of payments per year to 1 (which means a period is a year, and vice versa), the quantities involved in the calculation are:

| Cal. Key | Math Symbol | Explanation |
| --- | --- | --- |
| `N` | $n \cdot l$ | Number of periods |
| `I/Y` | $\frac{r}{n}$ | period interest rate |
| `PV` | $PV$ | Present value of all payments |
| `PMT` | $P$ | Payment amount per period |
| `FV` | $FV$ | Future value of all payments |

Notes:

1. The TVM calculation is always in the following fashion: given 4 out of the toal 5 values, and calculate the 5th missing value.

2. `N` is the total number of periods. For example, if the number of years is 5, and the number of payments per year is 4 (compounding quarterly), the total number of periods is 4*5=20.

3. `I/Y` is the effective interest rate for each period (compounding every period). For example, if the nominal annual interest rate is 12%, and the number of payments is 12 per year (compounding monthly), then the effective interest rate per period (month) is 12%/12=1%. Also, 10% interest rate should be inputted as 10, not 0.1.

4. There are two types of payments `PMT`: ordinary annuity and annuity due. Ordinary annuity makes payment at the end of each period, while the annuity due makes payments at the beginning of each period. The two modes on the calculator can be switched to each other using the key stroke `2nd` -> `BGN` -> `2nd` -> `SET`. When the annuity due mode is on, there is a `BGN` on top of the screen.

### Nominal vs Effective Interest Rate

Example: A monthly compounding loan with 12% annual (nominal) interest rate

- Nominal Interest Rate: periodic interest rate (used to calculate the incureed interest for each period) times the number of compounding periods per year. In the example above, the nominal interest rate is 12%, meaning that the monthly interest rate is 12%/12 = 1%. This 1% interest rate is used to calculate the interest incurred every month.

- Effective Interest Rate: this interest rate takes into account the compounding factor. One way to interpret this rate is that how much interest will be incurred during one year if one dollar was borrowed at the beginning of the year. The formula is as follows: ($i_e$ is the effective rate, $r$ is the nominal annual rate, and $n$ is number of periods per year)

$$
i_e = (1 + \frac{r}{n})^n -1
$$

### Discrete Compounding vs Continuous Compounding

- Discrete Compounding: compounded by day, week, month or quarter.

- Continuous Compounding: assuming the number of compounding period goes to infinity, i.e., the interest is generated continuously. The formula to compute continuous compounded cash flow is as follows: (e is the natrual number, r is the nominal annual interest rate, l is total number of years, FV stands for future value, PV stands for present value)

$$
FV = PV \times e^{rl}
$$

- Mathematical derivation of the continuous compounding formula:
    
$$
FV = lim_{n \rightarrow \infty} \big( PV \times (1+\frac{r}{n})^{nl} \big)
$$

Let $m=n/r$,
    
$$
\begin{align*}
& lim_{n \rightarrow \infty} \big( PV \times (1+\frac{r}{n})^{nl} \big) \\
=& lim_{m \rightarrow \infty} \big( PV \times (1+\frac{1}{m})^{mrl} \big) \\
=& lim_{m \rightarrow \infty} \big( PV \times [(1+\frac{1}{m})^{m}]^{rl} \big) \\
=& PV \times \big( lim_{m \rightarrow \infty}(1+\frac{1}{m})^{m} \big) ^{rl} \\
=& PV \times e ^{rl}
\end{align*}
$$

since the natural number $e$ can be defined as $e=lim_{m \rightarrow \infty}(1+\frac{1}{m})^{m}$.

### IRR and NPV for Uneven Cash Flow

The relationship between NPV and IRR (Internal Rate of Return) is:

$$
NPV = \sum^T_{t=1} \frac{C_t}{(1+IRR)^t} - C_0
$$

where:

$C_t$: net cash inflow during the period $t$;

$C_0$: total initial investment costs;

$IRR$: discount rate

$T$: total number of periods

In order to compute IRR, we set NPV=0 and then find a root to the equation, which in general doesn't have an analytical form. Numerical methods can be used.
